## comparission operators on arrays

**(<,>,==,!=,<=,>=)**

Any operations on array is performed element wise
when we use comparision operator on an array
we will get a boolean array

```python
import numpy as np
array1 = np.array([1, 2, 3])
array2 = np.array([3, 2, 1])

# less than operator
result1 = array1 < array2
print("array1 < array2:",result1)    # Output: [ True False False]

# greater than operator
result2 = array1 > array2
print("array1 > array2:",result2)    # Output: [False False  True]

# equal to operator
result3 = array1 == array2
print("array1 == array2:",result3)    # Output: [False  True False]
```

```python
# import random

# Function to roll the dice
def roll_dice():
    return random.randint(1, 6)

# Function to check if a player lands on a snake's head or ladder's bottom
def check_snake_or_ladder(position):
    snakes_and_ladders = {
        3: 17,
        7: 11,
        13: 5,
        19: 2,
        21: 9,
        27: 1,
        30: 15,
        34: 12,
        39: 24
    }
    return snakes_and_ladders.get(position, position)

# Main function to play the game
def play_snakes_and_ladders():
    print("Welcome to Snakes and Ladders!")
    
    # Collect player names
    player1_name = input("Enter Player 1's name: ")
    player2_name = input("Enter Player 2's name: ")

    player1_position = 0
    player2_position = 0

    while True:
        # Player 1's turn
        input(f"{player1_name}, press Enter to roll the dice.")
        dice_value = roll_dice()
        print(f"{player1_name} rolled a {dice_value}.")
        player1_position += dice_value
        player1_position = check_snake_or_ladder(player1_position)

        # Check if Player 1 has won
        if player1_position >= 40:
            print(f"{player1_name} wins!")
            break

        print(f"{player1_name} is at position {player1_position}.\n")

        # Player 2's turn
        input(f"{player2_name}, press Enter to roll the dice.")
        dice_value = roll_dice()
        print(f"{player2_name} rolled a {dice_value}.")
        player2_position += dice_value
        player2_position = check_snake_or_ladder(player2_position)

        # Check if Player 2 has won
        if player2_position >= 40:
            print(f"{player2_name} wins!")
            break

        print(f"{player2_name} is at position {player2_position}.\n")

# Start the game
play_snakes_and_ladders()
```
